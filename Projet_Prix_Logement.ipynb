### <font color= blue>Quelques préliminaires</font>

#### <font color= green>Bref revue des méthodes d'évaluation automatisées du prix d'un logement</font>

L'une des méthodes traditionnelles d'évaluation est le **modèle «comparables»**, une forme de **régression à k plus proches voisins**. Dans ce modèle, on suppose que la valeur de la propriété à évaluer est étroitement liée aux prix de vente de propriétés similaires dans la même zone. L'évaluateur doit ajuster le prix de vente de chaque comparable pour tenir compte des différences entre le sujet et le comparable. La valeur marchande du bien est déduite des prix de vente ajustés des comparables. Cependant, Cette approche dépend fortement de l'exactitude et de la disponibilité des données sur les transactions de vente. Une autre methode consiste a faire une **régression lineaire** dans lequel la variable dépendante est la valeur marchande d'une maison et les variables indépendantes sont les caractéristiques de la maison comme la taille, l'âge, le nombre de chambres, etc. De nombreux travaux appliquent des méthodes de régression au problème de l'estimation des prix immobiliers. Les modèles de régression linéaire supposent que la valeur marchande est une somme pondérée des caractéristiques du logement. Ils ne sont pas robustes aux valeurs aberrantes et ne peuvent pas traiter la non-linéarité dans les données. Un autre modèle utilisé pour l'estimation des prix est le **modèle de prix hédonique**. Elle est d'un emploi courant en économie et statistique. L'intuition, simple, est que le prix d'un bien(appartement, voiture, ordinateur...) ou d'un service(salaire)estfonction de ses caractéristiques. Par
exemple,le prix d'un logements expliquerait par sa surface,son nombre de pièces mais également par les infrastrures ou la pollution présentes dans son voisinage. La metode des prix hedonique suppose que la relation entre le prix et les variables indépendantes est une relation logarithmique non linéaire. D'autres approches pour l'estimation des prix comprennent **les réseaux de neurones**.

[bienici](https://www.bienici.com) utilise leur propre algorithmes pour l'estimation des prix immobiliers. Les caractéristiques de la maison, telles que la superficie en pieds carrés, l'emplacement ou le nombre de salles de bains, reçoivent des poids différents en fonction de leur influence sur les prix de vente des maisons dans chaque emplacement spécifique sur une période de temps spécifique, ce qui entraîne un **ensemble de règles d'évaluation** ou des modèles qui s'appliquent à générer une estimation du prix du bien immobilier.

#### <font color= green>Deep Learning : Convolutional neural networks (ConvNets)</font>

Un réseau de neurones convolutifs (**ConvNet/CNN**) est un algorithme d'apprentissage profond qui peut prendre une image d'entrée, attribuer de l'importance (poids et biais apprenables) à divers aspects/objets de l'image et être capable de différencier l'un de l'autre. Le prétraitement requis dans un **ConvNet** est bien inférieur à celui d'autres algorithmes de classification. Alors que dans les méthodes primitives, les filtres sont conçus à la main, avec suffisamment de formation, les **ConvNets** ont la capacité d'apprendre ces filtres/caractéristiques. Les réseaux de neurones convolutifs (**ConvNets**) ont atteint des performances de pointe sur des tâches telles que la reconnaissance d'images, la segmentation, la détection d'objets et la modélisation générative ces dernières années. Le récent regain d'intérêt pour les **ConvNets** a récemment entraîné l'apparition de nouvelles approches et architectures. L'un des premiers **ConvNets** influents était **LeNet**, qui était utilisé pour classer les chiffres manuscrits. Cependant, ce n'est que récemment que les **ConvNets** ont fourni des performances supérieures sur les tâches de classification et de détection d'objets par rapport aux méthodes précédentes telles que les fonctionnalités artisanales et les machines vectorielles de support. Le succès d'AlexNet ans la classification d'images dans le cadre du défi de reconnaissance visuelle à grande échelle **ImageNet** a ravivé l'intérêt pour les **ConvNets**. Les architectures de pointe telles que les **ResNets** et **DenseNet** permettent une meilleure information et un flux de gradient en utilisant des couches de saut ou en connectant différentes couches au sein du réseau.

```python
## Fonction de recuperation des liens des maisons d'un departements
def RecuperationLien(link,npage) : 
    lien_logement = []  
    
    for page in range(1,npage):
        lien=link+str(page)+"&mode=galerie"
        driver.get(lien)
        sleep(3)
        
        classes=driver.find_elements_by_class_name('details')
        
        for clas in classes : 
            lien_logement.append(clas.find_element_by_css_selector('a').get_attribute('href'))
        
    return lien_logement  



Id_global = []
Titre = []
Addresse = []
Prix = []
Prix_sur = []
Date_publication = []
Date_modification = []
Reference = []
Titre_description = []
Description = []
Surface = []
Nb_piece = []
Nb_chambre = []
Parking = []
Jardin = []
Ascenseur = []
Terrasse = []
Garage = []
    
def RecuperationOther(dab,emplacem) :
        
    with open('/content/drive/MyDrive/Projet Info 2020/base_lien/' + dab + '.csv','rb') as file : 
        db = pd.read_csv(file,usecols = ['id_house','liens'],encoding = 'utf-8')
        
    for i in range(0,db.shape[0]) : 
        print(i)
        
        driver.get(db['liens'].iloc[i])
        idx = db['id_house'].iloc[i]
        
        sleep(10)
        try : 
            prix =  driver.find_element_by_xpath('//div/div[3]/div[1]/div[1]/div[2]/div[1]/div[1]/div/div/span[1]').text
        except : 
            prix = ''
        titre = driver.title

        if 'à' in prix : 
            continue
        else :  
                try :
                    Adresse = driver.find_element_by_xpath("//h1/span").text
                except : 
                    Adresse = driver.find_element_by_xpath("//h1").text

                try : 
                    prices_surf =  driver.find_element_by_xpath('//div/div[3]/div[1]/div[1]/div[2]/div[1]/div[1]/div/div/span[2]').text
                except : 
                    prices_surf  = ''

                try :
                    publier = driver.find_element_by_xpath('//div/div[3]/div[2]/span[1]').text
                except : 
                    publier = ''

                try : 
                    modif = driver.find_element_by_xpath('//div/div[3]/div[2]/span[2]').text
                except : 
                    modif = ''

                try :
                    ref = driver.find_element_by_xpath('//div/div[3]/div[2]/span[3]').text
                except :
                    ref = ""

                try :
                    titre_descrip = driver.find_element_by_xpath('//div/div[3]/section[1]/h2').text
                except :
                    titre_descrip = ''

                try : 
                    descrip = driver.find_element_by_xpath('//div/div[3]/section[1]/div').text
                except : 
                    descrip = ''
      
                Id_global.append(idx)
                Titre.append(titre)
                Prix.append(prix)
                Addresse.append(Adresse)
                Prix_sur.append(prices_surf)
                Date_publication.append(publier)
                Date_modification.append(modif)
                Reference.append(ref)
                Titre_description.append(titre_descrip)
                Description.append(descrip)
                
                ### Recuperation des details sur la maison
                content = driver.find_elements_by_css_selector('div.labelInfo')
            
                tableau = []
                for j in range(len(content)) : 
                        tableau.append(content[j].text)
                
                Surface.append(next((x for x in tableau if bool(re.search("^\d{1,}[,]?\d{1,}?[ ]+m", x)) & bool(re.search(".*(?<!terrain)$",x))),''))
                Nb_piece.append(next((x for x in tableau if bool(re.search("^\d{1,}?[ ]pièce",x))),''))
                Nb_chambre.append(next((x for x in tableau if bool(re.search("^\d{1,}?[ ]chambre",x))),''))
                Parking.append(next((x for x in tableau if ('parking' in x.lower())),''))
                Jardin.append(next((x for x in tableau if ('jardin' in x.lower())),''))
                Ascenseur.append(next((x for x in tableau if ('ascenseur' in x.lower())),''))
                Terrasse.append(next((x for x in tableau if ('terrasse' in x.lower())),''))
                Garage.append(next((x for x in tableau if ('garage' in x.lower())),''))                    
                      
                    
                ### Recuperation des images lies a la maison                       
                images = driver.find_elements_by_css_selector('img')
    
                link = []
                pho_num = 0
                for each in images : 
                    try :
                        u = each.get_attribute('u')
                    except :
                        continue
                    if u == 'image' :
                        src = each.get_attribute('src')
                        pho_num = pho_num+1
                        if pho_num>len(images)/2 :
                            break
                        else : 
                            r = requests.get(src)
                            photo_id = idx+"_photo_"+str(pho_num)
                            empl = "/content/drive/MyDrive/Projet Info 2020/"+emplacem+"/"+photo_id+".jpg"
                            with open(empl,'wb') as f:
                                f.write(r.content)
                           

    dfinale = pd.DataFrame({'id_house':Id_global,'Addresse':Addresse,'Titre':Titre,'Prix':Prix,'Surface':Surface,'Nb_piece':Nb_piece,'Nb_chambre':Nb_chambre,'Parking':Parking,'Jardin':Jardin,
                            'Ascenseur':Ascenseur,'Terrasse':Terrasse,'Garage':Garage,'Prix_sur':Prix_sur,'Reference':Reference,'Date_publication':Date_publication,'Date_modification':Date_modification,
                            'Titre_description':Titre_description,'Description':Description})
    return dfinale
```

Le **comparateur de territoire** de l'INSEE (https://www.insee.fr/fr/statistiques/zones/1405599) donne pour chaque commune un certain nombre de données socio-économiques.
```python
def scrap(item):
    dict={"Nom de la commune (Code INSEE)":"","Population en 2017":"","Densité de la population (nombre d'habitants au km²) en 2017":"","Superficie en 2017, en km²":"","Nombre de ménages en 2017":"","Nombre total de logements en 2017":"","Médiane du revenu disponible par unité de consommation en 2017, en euros":"","Taux d'activité des 15 à 64 ans en 2017":"","Taux de chômage des 15 à 64 ans en 2017":""}
    url = "https://www.insee.fr/fr/statistiques/1405599?geo=COM-"+str(item)
    request_text = request.urlopen(url).read()
    page = bs4.BeautifulSoup(request_text, "lxml")
    commune=page.find('table').find('tr').findAll('th')[1]
    dict["Nom de la commune (Code INSEE)"]=commune.getText()
    for table in page.findAll("table") :
        for line in table.findAll("tr") :
            if line.find("td")==None or line.find("th")==None :
                pass
            else :
                if line.find("td").getText()=='' :
                    pass
                else :
                    if line.find("th").getText() in dict :
                        fig=float(line.find("td").getText().replace('\xa0','').replace(',','.').replace('–','-'))
                        dict[line.find("th").getText()]=fig
                    else :
                        pass
    return dict
```

```python
### Suppression des doublons
import hashlib
import matplotlib.pyplot as plt
import matplotlib.gridspec as gridspec
%matplotlib inline
import time
import numpy as np

def file_hash(filepath):
    with open(filepath, 'rb') as f:
        return md5(f.read()).hexdigest()
```

```python
file_list = os.listdir()
print(len(file_list))
```

```python
import hashlib, os
duplicates = []
hash_keys = dict()
for index, filename in  enumerate(os.listdir('.')):  #listdir('.') = current directory
    if os.path.isfile(filename):
        with open(filename, 'rb') as f:
            filehash = hashlib.md5(f.read()).hexdigest()
        if filehash not in hash_keys: 
            hash_keys[filehash] = index
        else:
            duplicates.append((index,hash_keys[filehash]))
```

```python
for file_indexes in duplicates[:30]:
    try:
    
        plt.subplot(121),plt.imshow(plt.imread(file_list[file_indexes[1]]))
        plt.title(file_indexes[1]), plt.xticks([]), plt.yticks([])

        plt.subplot(122),plt.imshow(plt.imread(file_list[file_indexes[0]]))
        plt.title(str(file_indexes[0]) + ' duplicate'), plt.xticks([]), plt.yticks([])
        plt.show()
    
    except OSError as e:
        continue
```
